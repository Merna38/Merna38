```python
import csv # comma seperated values
```

```python
csvfile = open('AIonlinejob.csv','w', encoding='utf-8')
writer =csv.writer(csvfile)
```

```python
writer.writerow(["Title_job","Company_name","Jobtype","Country"])
jobs=website.find_all("div",class_="css-pkv5jc")
for job in jobs:
    title_job = job.find("a",class_="css-o171kl").text
    company_name =job.find("a",class_="css-17s97q8").text
    jobtype =job.find("span",class_="css-1ve4b75").text
    country=job.find("span",class_="css-5wys0k").text
    writer.writerow([title_job ,company_name ,jobtype,country])
csvfile.close()   
```
